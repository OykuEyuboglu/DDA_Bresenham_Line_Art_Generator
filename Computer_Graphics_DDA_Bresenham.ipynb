```python
import pygame
import sys
import random
import math
```

```python
pygame.init()

width = 800
height = 600

screen = pygame.display.set_mode((width, height))
pygame.display.set_caption("DDA and Bresenham Art")

clock = pygame.time.Clock()
```

```python
def dda_line(x1, y1, x2, y2, color):

    dx = x2 - x1
    dy = y2 - y1

    steps = int(max(abs(dx), abs(dy)))

    x_inc = dx / steps
    y_inc = dy / steps

    x = x1
    y = y1

    for i in range(steps):
        screen.set_at((int(x), int(y)), color)
        x += x_inc
        y += y_inc
```

```python
def draw_sparkles():

    for i in range(80):

        x = random.randint(0,800)
        y = random.randint(0,600)

        screen.set_at((x,y), (255,255,255))  

        screen.set_at((x+1,y), (200,200,200))
        screen.set_at((x-1,y), (200,200,200))
        screen.set_at((x,y+1), (200,200,200))
        screen.set_at((x,y-1), (200,200,200))
```

```python
def bresenham_line(x1, y1, x2, y2, color):

    dx = abs(x2-x1)
    dy = abs(y2-y1)

    sx = 1 if x1<x2 else -1
    sy = 1 if y1<y2 else -1

    err = dx - dy

    while True:

        screen.set_at((x1,y1), color)

        if x1==x2 and y1==y2:
            break

        e2 = 2 * err

        if e2 > -dy:
            err -= dy
            x1 += sx

        if e2 < dx:
            err += dx
            y1 += sy
```

```python
def draw_random_art():

    for i in range(50):

        x1 = random.randint(0,800)
        y1 = random.randint(0,600)

        x2 = random.randint(0,800)
        y2 = random.randint(0,600)

        color = (
            random.randint(0,255),
            random.randint(0,255),
            random.randint(0,255)
        )

        if i % 2 == 0:
            dda_line(x1,y1,x2,y2,color)
        else:
            bresenham_line(x1,y1,x2,y2,color)
```

```python
running = True

while running:

    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            running = False

    screen.fill((0,0,0))
    
    draw_sparkles()
    draw_random_art()

    pygame.display.update()

    clock.tick(2)

pygame.quit()
```
